```python
# Eliminar la palabra "bhp", "CC", "Kmpl" ,  de la columnas 


Datos['max_power'] = Datos['max_power'].str.replace('bhp', '')
Datos['engine'] = Datos['engine'].str.replace('CC', '')
Datos['mileage'] = Datos['mileage'].str.replace('kmpl', '')
Datos['mileage'] = Datos['mileage'].str.replace('km/kg', '')


#Convertir a valor numerico mileage, engine, Max_power que seran tenidos en cuenta para el modelo 
Datos['mileage'] = pd.to_numeric(Datos['mileage'], errors='coerce')
Datos['engine'] = pd.to_numeric(Datos['engine'], errors='coerce')
Datos['max_power'] = pd.to_numeric(Datos['max_power'], errors='coerce')



```

```python
#Convertir a valor numerico mileage, engine, Max_power para poder implementarlos en el modelo
Datos['mileage'] = pd.to_numeric(Datos['mileage'], errors='coerce')
Datos['engine'] = pd.to_numeric(Datos['engine'], errors='coerce')
Datos['max_power'] = pd.to_numeric(Datos['max_power'], errors='coerce')
```

```python
Datos.info()
```

```python

# utilizamos ciclo for  para que muestre las columnas del dataset que contienen valores vacíos.
for column in Datos.columns:
    datos_faltantes = Datos[column].isna().sum()
    print("Cantidad de valores faltantes en {}: {}".format(column, datos_faltantes))
```

```python
# Calcula la moda de "mileage" , "engine" y "max_power"
moda_mileage = Datos["mileage"].mode()[0]
moda_engine = Datos["engine"].mode()[0]
moda_max_power=Datos["max_power"].mode()[0]

print(moda_mileage )
print(moda_engine)
print(moda_max_power)

# Rellena los valores vacíos con la moda
Datos["mileage"].fillna(moda_mileage, inplace=True)
Datos["engine"].fillna(moda_engine, inplace=True)
Datos["max_power"].fillna(moda_engine, inplace=True)
```

El MAE mide la magnitud promedio de los errores absolutos entre las predicciones y los valores reales.
 MAE es de aproximadamente 121,119.12. Esto significa que, en promedio, las predicciones del modelo difieren en alrededor de 121,119.12 unidades monetarias (por ejemplo, dólares) del valor real de venta.
MSE (Error Cuadrático Medio):
El MSE es una medida de la dispersión de los errores al cuadrado.
 el MSE es de aproximadamente 26,060,763,054.90. Cuanto mayor sea el MSE, mayor será la variabilidad de los errores.
RMSE (Raíz del Error Cuadrático Medio):
El RMSE es simplemente la raíz cuadrada del MSE. El RMSE es de aproximadamente 161,433.46. Representa la dispersión promedio de los errores en la misma unidad que la variable objetivo (precio de venta).
Análisis General:
El MAE y el RMSE indican que el modelo tiene errores significativos en las predicciones.
